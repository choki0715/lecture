# 판다스 불러오기 및 Series 자료구조

```python
import pandas as pd

ss_1 = pd.Series([100,88,56,72,92], index=['학생1','학생2','학생3','학생4','학생5'])

# 전체 데이터 보기
print(ss_1)
```

```python

print(ss_1.values)
print(ss_1.index)

# 특정 데이터 보기
print(ss_1.iloc[2])
print(ss_1.loc['학생3'])

dict_1 = {'학생6': 68, '학생7':74}
ss_2 = pd.Series(dict_1)

print(ss_2)
```

# DataFrame 자료구조

```python
dict_2 = {'이름':['준호', '명수', '창훈'], '영어성적':[100, 80, 73], '수학성적':[80, 90, 53]}
df = pd.DataFrame(dict_2)
print(df)
print(df.values)
print(df.index)
print(df.columns)
```

```python

print(df.iloc[1,1])
print(df.loc[1,'영어성적'])
```

```python
# 선택적인 특성 선정
feature = df[['total_rooms', 'population', 'households']]
short_df = feature.head()
short_df
```

```python
# 컬럼 추가
short_df.insert(0, 'location', ['서울', '서울', '경기', '대전', '경기'])
short_df.head()
```

```python
# 로우 추가
short_df.loc[5] = ['광주', 1234, 150, 320]
short_df
```

```python
# concat을 이용한 로우 추가
cols = ['location',	'total_rooms',	'population',	'households']
new_row = pd.DataFrame([['부산', 10, 20, 30]], columns=cols)

short_df = pd.concat([short_df, new_row], ignore_index=True)
short_df
```

```python
idx = 6
short_df = short_df.drop([short_df.index[idx]])
short_df
```

```python
data = short_df.values
data
```

```python
data_mean = data.mean(axis=0)
data_std = data.std(axis=0)

print('data_mean', data_mean)
print('data_std', data_std)
```

```python
short_df = short_df.drop(['location'], axis=1)

data = short_df.values
data
```

```python
data_mean = data.mean(axis=0)
data_std = data.std(axis=0)

print('data_mean', data_mean)
print('data_std', data_std)
```
