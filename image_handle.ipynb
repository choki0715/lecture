#OpenCV (cv2)를 이용한 이미지 입력

```python
import cv2
from google.colab.patches import cv2_imshow

# Google Colab 세션 관리를 위해 cv2_imshow로 대체
# cv2.imshow(img)
# cv2.waitKey(0)

# 파일 위치 경로 지정하기
file_path = '/content/drive/MyDrive/'

# 파일명
celebs = 'face.jpg'
choki = 'choki.jpg'
cat = 'cat.png'

#read jpg, png 파일
celeb_jpg = cv2.imread(file_path + celebs)
choki_jpg = cv2.imread(file_path + choki)
cat_png = cv2.imread(file_path + cat, cv2.IMREAD_UNCHANGED)

# 이미지 array 차원
print('JPG format: ', choki_jpg.shape)
print('PNG format: ', cat_png.shape)

# 이미지 보기
cv2_imshow(cv2.resize(celeb_jpg, dsize=(400,220)))
cv2_imshow(cv2.resize(choki_jpg, dsize=(200,190)))
cv2_imshow(cv2.resize(cat_png, dsize=(200,200)))
```

```python
# OpenCV는 BGR 순으로 구성

celeb_jpg = cv2.imread(file_path + celebs)

celeb_jpg[:,:,0] = 0    # B(lue)
celeb_jpg[:,:,1] = 0    # G(reen)
#celeb_jpg[:,:,2] = 0    # R(ed)

cv2_imshow(cv2.resize(celeb_jpg, dsize=(400,250)))
```

#PyTorch에서 이미지 처리하기

```python
# Import the required libraries
import torch
import torchvision
from torchvision.io import read_image
import torchvision.transforms as T

# 파일 위치 경로 지정하기
file_path = '/content/drive/MyDrive/'

# 파일명
celebs = 'face.jpg'
choki = 'choki.jpg'
cat = 'cat.png'

# JPEG image 일기
img = read_image(file_path + celebs)

# check if input image is a PyTorch tensor
print("Is image a PyTorch Tensor:", torch.is_tensor(img))
print("Type of Image:", type(img))

# size of the image
print('torch tensor of image: ', img.size())

```
