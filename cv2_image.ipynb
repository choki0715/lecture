```python
import cv2 
import numpy as np
from google.colab.patches import cv2_imshow

img = cv2.imread('./drive/MyDrive/hollywood.jpeg')
#img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
r, g, b = cv2.split(img)

height, width, channel = img.shape
zero = np.zeros((height, width, 1), dtype=np.uint8)
bgz = cv2.merge((b, zero, zero))

nr = r/255.
ng = g/255.
nb = b/255.

cv2_imshow(img)
cv2_imshow(bgz)
```
